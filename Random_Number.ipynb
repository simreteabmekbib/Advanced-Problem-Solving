```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(N):
    x = np.random.uniform(0, 1, N)
    y = np.random.uniform(0, 1, N)
    inside_circle = x**2 + y**2 <= 1
    pi_estimate = 4 * np.sum(inside_circle) / N
    return pi_estimate

# Testing for different N values
N_values = [1, 10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**8]
pi_estimates = [estimate_pi(N) for N in N_values]

# Plotting convergence
plt.plot(N_values, pi_estimates, marker='o', label='Estimated π')
plt.axhline(np.pi, color='r', linestyle='--', label='True π')
plt.xscale('log')
plt.xlabel('Number of Points (N)')
plt.ylabel('Estimated π')
plt.legend()
plt.title('Convergence of π Estimation')
plt.show()
```

```python
def monte_carlo_integration(f, a, b, N):
    x = np.random.uniform(a, b, N)
    return (b - a) * np.mean(f(x))

# Single-dimensional integral
f = lambda x: np.exp(-x**2)
true_value = 0.746824  # Precomputed value of the integral
N_values = [1, 10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**8]
integral_estimates = [monte_carlo_integration(f, 0, 1, N) for N in N_values]

# Plotting convergence
plt.plot(N_values, integral_estimates, marker='o', label='Estimated Integral')
plt.axhline(true_value, color='r', linestyle='--', label='True Value')
plt.xscale('log')
plt.xlabel('Number of Points (N)')
plt.ylabel('Estimated Integral')
plt.legend()
plt.title('Convergence of Monte Carlo Integration')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f_2d(x, y):
    return np.exp(-(x**2 + y**2))

# Monte Carlo integration for multidimensional functions
def monte_carlo_integration_2d(num_points):
    # Generate random points (x, y) in the range [0, 1] x [0, 1]
    x_random = np.random.uniform(0, 1, num_points)
    y_random = np.random.uniform(0, 1, num_points)

    # Evaluate the function at the random points
    f_values = f_2d(x_random, y_random)

    # Estimate the integral as the mean of the function values over the domain
    integral_estimate = np.mean(f_values)
    return integral_estimate

# Experiment with increasing numbers of points
num_points_list = [1, 10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 10**8]
integral_estimates = []

# Compute the integral for each number of points
for num_points in num_points_list:
    estimate = monte_carlo_integration_2d(num_points)
    integral_estimates.append(estimate)

# Plot the convergence
plt.figure(figsize=(8, 6))
plt.plot(num_points_list, integral_estimates, marker='o', label='Estimated Integral')
plt.axhline(y=0.557746, color='r', linestyle='--', label='True Value (0.557746)')
plt.xscale('log')
plt.xlabel('Number of Points (N)')
plt.ylabel('Estimated Integral')
plt.title('Convergence of Multidimensional Integral Estimation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define N values (e.g., powers of 10 for better visualization on log-log scale)
N_values = [1, 10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000]

# Error analysis for estimating π
errors_pi = []
for N in N_values:
    estimates = [estimate_pi(N) for _ in range(10)]
    error = np.std(estimates) / np.sqrt(N)
    errors_pi.append(error)

# Error analysis for 1D integration
errors_1d = []
for N in N_values:
    estimates = [monte_carlo_integration(f, 0, 1, N) for _ in range(10)]
    error = np.std(estimates) / np.sqrt(N)
    errors_1d.append(error)

# Error analysis for 2D integration
errors_2d = []
for N in N_values:
    estimates = [monte_carlo_integration_2d(N) for _ in range(10)]
    error = np.std(estimates) / np.sqrt(N)
    errors_2d.append(error)

# Plot all errors on a log-log scale
plt.figure(figsize=(10, 6))
plt.plot(N_values, errors_pi, marker='o', label='Pi Estimation Error')
plt.plot(N_values, errors_1d, marker='s', label='1D Integration Error')
plt.plot(N_values, errors_2d, marker='^', label='2D Integration Error')
plt.plot(N_values, [1/np.sqrt(N) for N in N_values], linestyle='--', label='Theoretical Error (1/sqrt(N))')
plt.xlabel('Number of Points (N)')
plt.ylabel('Error')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title('Error Analysis for Monte Carlo Methods')
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```
